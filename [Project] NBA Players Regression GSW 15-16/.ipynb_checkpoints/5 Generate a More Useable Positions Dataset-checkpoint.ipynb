## Convert the Ultra Pbp Positions to a more useable format for regression.

### Functions

```python
from bs4 import BeautifulSoup
import requests
from HTMLParser import HTMLParser
from datetime import datetime
import hashlib
import requests
import re

import numpy as np
import pandas as pd

class MLStripper(HTMLParser):
    def __init__(self):
        self.reset()
        self.strict = False
        self.convert_charrefs= True
        self.fed = []
    def handle_data(self, d):
        self.fed.append(d)
    def get_data(self):
        return ''.join(self.fed)

def strip_tags(html):
    s = MLStripper()
    s.feed(html)
    return s.get_data()


# In[58]:
from selenium import webdriver
import urllib2
import MySQLdb
```

### Read the files

```python
positions_DS = pd.read_csv("Pbp Data Positions.csv", sep=",")
positions_DS.head()
```

```python
positions = []
dataset = pd.DataFrame(columns=('id','SFnum_G', 'PFnum_G', 'Cnum_G', 'SGnum_G', 'PGnum_G', 'SFnum_O', 'PFnum_O', 'Cnum_O', 'SGnum_O', 'PGnum_O', 'secs', 'score_GSW', 'score_OPP', 'sG-sO','game'))

for i,row in positions_DS.iterrows():
    dict_pos_GSW = {'SF':0, 'PF':0, 'C':0, 'SG':0, 'PG':0}
    dict_pos_OPP = {'SF':0, 'PF':0, 'C':0, 'SG':0, 'PG':0}

    
    id_val = []
    end_values = []
    #iterate per row
    for j in range(0,17):
        
        if j == 0:
            id_val.append(int(row[j]))
        elif j >= 1 and i <= 5:
            dict_pos_GSW[row[j]] += 1
        elif j >= 6 and j <= 10:
            dict_pos_OPP[row[j]] += 1
        elif j != 15:
            end_values.append(int(row[j]))
        else:
            end_values.append(row[j])
    #record the
    record = id_val + [dict_pos_GSW['SF'], dict_pos_GSW['PF'], dict_pos_GSW['C'], dict_pos_GSW['PG'], dict_pos_GSW['SG']]
    record = record + [dict_pos_OPP['SF'], dict_pos_OPP['PF'],dict_pos_OPP['C'],dict_pos_OPP['PG'],dict_pos_OPP['SG']] + end_values
    print record

    temp = pd.DataFrame([record], columns=('id','SFnum_G', 'PFnum_G', 'Cnum_G', 'SGnum_G', 'PGnum_G', 'SFnum_O', 'PFnum_O', 'Cnum_O', 'SGnum_O', 'PGnum_O', 'secs', 'score_GSW', 'score_OPP', 'sG-sO','game'))
    frames = [dataset, temp]
    dataset = pd.concat(frames, ignore_index = True)

    
```

```python
dataset.to_csv("Ultra Pbp Positions for Regression.csv", sep=',')
```
