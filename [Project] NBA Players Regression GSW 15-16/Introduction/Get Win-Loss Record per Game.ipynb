```python
from bs4 import BeautifulSoup
import requests
from HTMLParser import HTMLParser
from datetime import datetime
import hashlib
import requests
import re

import numpy as np
import pandas as pd

class MLStripper(HTMLParser):
    def __init__(self):
        self.reset()
        self.strict = False
        self.convert_charrefs= True
        self.fed = []
    def handle_data(self, d):
        self.fed.append(d)
    def get_data(self):
        return ''.join(self.fed)

def strip_tags(html):
    s = MLStripper()
    s.feed(html)
    return s.get_data()


# In[58]:
from selenium import webdriver
import urllib2
import MySQLdb
```

```python
def get_name_teams(link):
    
    a = link.split("https://en.wikipedia.org/wiki/2015%E2%80%9316_")[1]
    b = a.split("_season")[0]
    c = " ".join(b.split("_"))
    
    return c
```

## Scrape the list of urls from the main NBA 2015-2016 season.

```python
wiki_url = "https://en.wikipedia.org/wiki/2015%E2%80%9316_NBA_season"

r1 = urllib2.urlopen(wiki_url)
soup = BeautifulSoup(r1,'html.parser')
tables = soup.findAll('table', {'style':'border-collapse: collapse; padding: 0; border: 0; background:transparent; width:100%;'})

base_url_wiki = u'https://en.wikipedia.org'
```

```python
teams_links = []


raw_links =  tables[0].find_all('a', href=True) 
print len(raw_links)

for i in raw_links:
    link = i['href']
    
    #what we want
    if "2015%E2%80%9316" in i['href']:
        suffix = i['href']
        the_link = base_url_wiki + suffix
        teams_links.append(the_link)
        print the_link

print len(teams_links)
```

```python
#get all the URLs for scraping!
#url composition: "http://www.nba.com/games/" + <date of game from wiki> + "//" + <team A AT team B>  + "//" + "/gameinfo.html"

#get url 
real_reg_season_soups = []
teams = []
for i2,link in enumerate(teams_links[:]):
    
    print i2, link,
    #use beautiful soup to parse more conveniently
    r1 = urllib2.urlopen(link)
    soup = BeautifulSoup(r1, 'html.parser')

    #losses = soup.findAll('tr', {'style': 'background:#fbb;'})
    #wins = soup.findAll('tr', {'style': 'background:#bfb;'})
    wiki_table = soup.findAll('table', {'style':'width:90%; clear:both; margin:1.5em auto; text-align:center;'})

    ilist = []
    for i in wiki_table:
        if "2015–16 game log" in i.encode("utf-8"):
            ilist.append(i)
    print len(ilist)
    real_reg_season_soups.append(ilist[-1])
            
    #get the relevant tables; i.e. the regular season sched and playoffs sched. We need these tables to generate the links
    teams.append(get_name_teams(link))
    print get_name_teams(link)
```

```python
df = pd.DataFrame(columns=('Team', 'Game Number', 'Team Score', 'Opponent Score', 'Win'))

for i,soup in enumerate(real_reg_season_soups[:]):
    games = soup.findAll('a', {'rel':'nofollow'})
    
    #remove the last item
    temp = games.pop(-1)
    
    for j,k in enumerate(games[:]):
        
        game_num = j + 1
        win = k.text.split(" ")[0]
        if win == 'W':
            win = 1
        else:
            win = -1
        split_text = k.text.split(" ")[1]
        split_text = split_text.split(u'\u2013')
        score_team = split_text[0]
        score_opp = split_text[1]
        record = [teams[i], game_num, score_team, score_opp, win]
        temp_df = pd.DataFrame([record],columns=('Team', 'Game Number', 'Team Score', 'Opponent Score', 'Win'))
        frames = [df, temp_df]
        df = pd.concat(frames, ignore_index = True)
```
